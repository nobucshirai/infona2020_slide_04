**III-B.** (2) 2分法により $f(x)$ の根を求めるプログラムを作成して問1で求めた回数繰り返し、繰り返し操作後の解の存在する $x$ の区間の両端の値を10進11桁以降を切り捨てて求めよ。またその区間の両端の値の平均値を求め根の近似値として10進11桁以降を切り捨てて求めよ。作成したプログラムも提出すること。プログラミング言語は問わない。


以下Python3を用いたプログラム例を示す。

```
# Pythonのバージョン確認
import sys
print(sys.version)
```

```
import math
def f(x):
    return 2 - math.exp(x)
```

```
a = 0; b = 1
for i in range(1,27):
    #c = (a+b)/2
    c = a + (b-a)/2 # 平均値の計算を旨くする
    if f(a) * f(c) > 0:
        a = c
    else:
        b = c

print()
print("解の存在する区間幅 [{:.15f}, {:.15f}]".format(a, b))
c = a + (b-a)/2
print("根の近似値 {:.15f}".format(c))
print("残差 {:.15f}".format(abs(f(c))))
```
